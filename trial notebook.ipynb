```python
def retrievestocklinktwo(asymbol):
    mydata = qd.get('WIKI/' + str(asymbol), start_date = '2017-04-01', end_date = '2017-04-30')
    return mydata
```

```python
df= pd.DataFrame(retrievestocklinktwo('GOOG'))
dfreset = df.reset_index()
dfreset
```

```python
def makegraph(asymbol, symboldf):
    sf = figure(title='Stock Price for '+'GOOG'+' during April 2017')
    sf.xaxis.axis_label = 'April 2017'
    sf.yaxis.axis_label = 'Closing Price'
    sf.line(x=np.array(dfreset['Date'].values, dtype=np.datetime64), y=np.array(dfreset['Close'].values),line_width=2, legend='Close')
    return sf
```

```python
@app.route('/stockgraph', methods=['POST'])
def stockgraph():
    app.vars['stocksymbol'] = request.form['stocksymbol']
    script, div = components(sf)
    return render_template('stockgraph.html', script=script, div=div)
```
